# TITLES LOWER

```python
# Asegrate de que df_mh es un DataFrame antes de realizar cambios
df_mh.columns = df_mh.columns.str.lower()
print(df_mh.columns)  # Para ver las columnas en minsculas
```

# HOMOGENISATION UNIQUE VALUES

```python
def categorize_gender(gender):
    gender = gender.lower()  # Convertir todo a minsculas para simplificar comparaciones

    # Definir las categoras
    male_labels = [
        'male', 'm', 'man', 'cis male', 'guy', 'male-ish', 'male (cis)', 'maile', 'mal', 'malr', 'mail', 
        'guy (-ish) ^_^', 'make', 'cis man', 'man', 'msle', 'male-ish'
    ]
    female_labels = [
        'female', 'f', 'woman', 'cis female', 'femake', 'femail', 'female (cis)', 'female (trans)', 'cis-female/femme',
        'woman', 'female (cis)', 'cis woman'
    ]
    lgbt_labels = [
        'trans woman', 'trans-female', 'female (trans)', 'non-binary', 'fluid', 'queer', 'genderqueer', 'agender', 
        'androgyne', 'neuter', 'enby', 'male leaning androgynous', 'something kinda male?', 'a little about you',
        'ostensibly male, unsure what that really means', 'queer/she/they', 'genderqueer', 'nah', 'all', 'fluid', 'p'
    ]

    # Revisar y categorizar
    if any(label in gender for label in male_labels):
        return 'Male'
    elif any(label in gender for label in female_labels):
        return 'Female'
    elif any(label in gender for label in lgbt_labels):
        return 'Queer/Dissidents'
    else:
        return gender  # Para otros valores raros o no categorizados

# Aplicar la funcin a la columna Gender
df_mh['gender'] = df_mh['gender'].apply(categorize_gender)

# Ver los resultados
df_mh['gender'].value_counts()

```

# Imputacin Basada en Proporciones.
### Proporciones de Frecuencia: puedes calcular las proporciones de cada categora y usar esas proporciones para realizar la imputacin. 

```python
# Calcular las proporciones de cada categora
proportions = df_mh['work_interfere'].value_counts(normalize=True)

# Generar imputaciones basadas en las proporciones
n_nans = df_mh['work_interfere'].isnull().sum()
imputed_values = np.random.choice(proportions.index, size=n_nans, p=proportions.values)

# Asignar los valores imputados a los nulos
df_mh.loc[df_mh['work_interfere'].isnull(), 'work_interfere'] = imputed_values

# Verificar el resultado
print(df_mh['work_interfere'].value_counts())
```
