```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("../Files/bank-additional-clean-nonulls.csv")
```

```python
# Gráficos de variables numéricas

# Histograma en seaborn:

sns.histplot(data=df, x='age', color='g', bins=20, kde=True)

# el KDE es la tendencia de distribución de los datos
```

```python
# Histograma en MatPotLib:

plt.hist(data=df, 
         x='age',
         bins=20,
         edgecolor='black',
         range=[0,100])

# Poner datos a los gráficos:

plt.xlabel('Edad')
plt.ylabel('Cantidad')
```

```python
# Dispersión:

sns.scatterplot(data=df,
                x='age',
                y='pdays',
                color='g',
                marker='v')

plt.xlabel('Edad', fontsize=15)
```

```python
# Regresión:

sns.regplot(data=df, x='pdays', y='age', line_kws={'color':'red'}, scatter_kws={'color':'black'})
```

```python
# Podemos representar varios gráficos en una misma figura o imagen

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(40,20)) # creamos los espacios para poner los gráficos

# seaborn
sns.regplot(data=df,
                x='pdays',
                y='age',
                color='black',
                line_kws={'color':'red'},
                scatter_kws={'color':'blue'},
                ax=axs[0,1])

# matpotlib
axs[0,0].hist(data=df,
              x='age',
              bins=20,
              color='g',
              edgecolor='black',
              range=[0,100])
```

```python
df['age'].mean()
```

```python
# BoxPlot

sns.boxplot(data=df,
            x='age',
            color='green'
            )
```

```python
# Mismo gráfico pero con Matplotlib

plt.boxplot(data=df, x='age')
```

```python
# Violin Pilot
sns.violinplot(data=df,
               y='age',
               color='green')
```

```python
# Variables categóricas

df['education'].unique() # comprobamos los valores únicos de una variable categórica como educación.

# Countplot: cuenta las ocurrencias de cada valor.

sns.countplot(data=df,
              x='education',
              color='green')

plt.xticks(rotation=90) # con esto le damos espaciado para que las columnas se muestren de la manera más correcta visualmente
```

```python
# Pieplot - grafiquito de tartas

df_education = df.groupby('education')['id'].count().reset_index()
# cuantos elementos hay por cada elemento educativo. 
df_education
```

```python
plt.pie('id', labels='education',data=df_education, autopct='%1.1f%%', startangle=90);
```

```python
# Analizar variables numéricas y categóricas:

# Barplot

sns.barplot(data=df,
            x='marital',
            y='income',
            palette='magma')
```
