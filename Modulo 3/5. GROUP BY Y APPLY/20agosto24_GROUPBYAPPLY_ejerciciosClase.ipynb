```python
# Librerías a utilizar

import pandas as pd
import numpy as np

# Configuración

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('bank-additional.csv')
```

```python
# aqui podemos ver los valores únicos de la columna educación, nos indica que es una variable de tipo categórica, podemos hacer agrupación 
data['education'].unique()
```

```python
data.groupby('education').mean(numeric_only=True) #media agrupando por el valor de educacion
```

```python
# Podemos agrupar por varias columnas

data.groupby(['education','marital']).mean(numeric_only=True)
```

```python
# cada una de las educaciones pero solamente nos calcula la media de edad.
data.groupby(['education','marital'])['age'].mean(numeric_only=True)
```

```python
# voy a contar cuantas personas hay de cada grupo

data.groupby(['education','marital'])['id_'].count().reset_index() # el reset index es para que no se pierda la estructura de DataFrame
```

```python
# Funciones de agregación: sum, mean, median, min, max, count

# Función agg

data.groupby('education')['age'].agg(['mean','median','min','max','count'])
```

```python
data.describe()
```

```python
# APPLY

# Apply se utiliza para aplicar una función a cada columna o fila de un DataFrame

# PASO 1: Añadir una nueva columna para categorizar la edad de las personas.
# menos de 30, joven; entre 30 y 50, adulta y más de 50: mayor; dentro de una función

# PASO 2: aplicar la función a la columna 'age' del DataFrame.

def agrupar_edad(num):

    if num < 30:
        res="Joven"
    elif num >= 30 and num <= 50:
        res="Adulta"
    else:
        res="Mayor"
    return res

data['age_group'] = data['age'].apply(agrupar_edad)
```

```python
data.sample(2)
```

```python
data.insert(7, "age_cat", data["age"].apply('age_group'), allow_duplicates=False)
```

```python
# Map_ aplicar una función a cada valor de una columna

data['loan'] = data['loan'].map({0.0:"No",1.0:"Yes"})
```

```python
## Replace - reemplazar

data['pdays'] = data['pdays'].replace(999, np.nan)
```
