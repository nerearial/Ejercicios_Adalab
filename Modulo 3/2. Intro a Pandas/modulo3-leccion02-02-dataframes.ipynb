## `pd.read_csv()`

La sintaxis b치sica de `pd.read_csv` es la siguiente:

```python
pd.read_csv(ruta_al_fichero , sep=',', delimiter=None, header='infer', names=None, index_col=None, dtype=None)
```

Donde:
- `ruta_al_fichero`: Es la ruta del archivo CSV que deseas leer.

- `sep` (opcional): Es el separador utilizado en el archivo CSV para separar los valores. Por defecto, es una coma (',').

- `delimiter` (opcional): Es un alias para `sep`. Puedes usar `delimiter` en lugar de `sep` si prefieres.

- `header` (opcional): Indica la fila del archivo CSV que se utilizar치 como encabezado de columnas. Puede tener los siguientes valores:
  - `None`: No hay encabezado y los datos se leer치n sin nombres de columna.
  - `'infer'`: Intentar치 inferir autom치ticamente los nombres de columna del archivo.
  - Un n칰mero *integer* que especifica la fila del archivo que se utilizar치 como encabezado de columna (por ejemplo, `header=0`).


- `names` (opcional): Es una lista opcional de nombres para las columnas. Si se proporciona, reemplazar치 los nombres inferidos o los nombres de columna del archivo.

- `index_col` (opcional): Especifica la columna (o columnas) que se utilizar치 como 칤ndice del DataFrame resultante. Puede ser un n칰mero *integer*, una cadena con el nombre de la columna o una lista de columnas.

- `dtype` (opcional): Permite especificar el tipo de datos de las columnas. Puede ser un diccionario que mapee nombres de columna a tipos de datos o un solo tipo de datos para aplicar a todas las columnas.

Estos son solo algunos de los par치metros m치s comunes utilizados en `pd.read_csv`. La funci칩n `pd.read_csv` tambi칠n admite muchos otros par치metros, como `skiprows`, `na_values`, `parse_dates` y m치s. Siempre podemos consultar la documentaci칩n oficial de este m칠todo [aqu칤](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) para obtener informaci칩n m치s detallada sobre estos par치metros y sus opciones.

````python
df_jobs = pd.read_csv("jobs.csv")
df_jobs.head(1)
````

Como la propia definici칩n indica, nuestras columnas deben ir separadas por comas. A veces, puede ocurrir que no vengan separados por comas. Como en el ejemplo que vemos a continuaci칩n:

- `index_col` (opcional): Especifica la columna (o columnas) que se utilizar치 como 칤ndice del DataFrame resultante. Puede ser un n칰mero *integer*, una cadena con el nombre de la columna o una lista de columnas.

- `dtype` (opcional): Permite especificar el tipo de datos de las columnas. Puede ser un diccionario que mapee nombres de columna a tipos de datos o un solo tipo de datos para aplicar a todas las columnas.

Adem치s de estos par치metros, `pd.read_excel` tambi칠n admite otros par치metros opcionales, como `skiprows`, `na_values`, `usecols` y m치s. ## `pd.read_json()`

La sintaxis b치sica de `pd.read_json` es la siguiente:

```python

```

Donde:
- `ruta_al_fichero`: Es la ruta del archivo JSON que deseas leer.

- `orient` (opcional): Especifica la orientaci칩n del archivo JSON. Puede tener los siguientes valores:

  - `'columns'` (por defecto): Se espera que el archivo JSON tenga un formato de columna.

  - `'index'`: Se espera que el archivo JSON tenga un formato de 칤ndice.

  - `'records'`: Se espera que el archivo JSON tenga un formato de registros.

  - `'split'`: Se espera que el archivo JSON tenga un formato dividido.

  - `'values'`: Se espera que el archivo JSON contenga solo valores sin ninguna etiqueta de columna o 칤ndice.

- `typ` (opcional): Indica el tipo de objeto a crear. Puede tener los siguientes valores:

  - `'frame'` (por defecto): Crea un objeto DataFrame.

  - `'series'`: Crea un objeto Series.

- `dtype` (opcional): Permite especificar el tipo de datos de las columnas. Puede ser un diccionario que mapee nombres de columna a tipos de datos o un solo tipo de datos para aplicar a todas las columnas.

- `convert_axes` (opcional): Indica si las etiquetas de los ejes deben convertirse en 칤ndices o nombres de columna. Por defecto, es `True`.

- `convert_dates` (opcional): Indica si se deben convertir las cadenas de fecha y hora en objetos de fecha y hora. Por defecto, es `True`.

- `keep_default_dates` (opcional): Especifica si se deben mantener las fechas predeterminadas al convertir cadenas de fecha y hora. Por defecto, es `True`.

- `numpy` (opcional): Indica si los datos deben devolverse como una matriz NumPy en lugar de un objeto DataFrame. Por defecto, es `False`.

- `precise_float` (opcional): Indica si se deben utilizar n칰meros de punto *float* precisos en lugar de valores de punto *float* nativos de Python. Por defecto, es `False`.

- `date_unit` (opcional): Especifica la unidad de fecha y hora si se deben convertir las cadenas de fecha y hora. Puede ser `'s'` para segundos o `'ms'` para milisegundos.

- `encoding` (opcional): Permite especificar la codificaci칩n del archivo JSON si no se puede inferir autom치ticamente.

- `lines` (opcional): Indica si el archivo JSON contiene m칰ltiples objetos JSON en l칤neas separadas en lugar de un solo objeto JSON.

Estos son solo algunos de los par치metros m치s comunes utilizados en `pd.read_json`. La funci칩n `pd.read_json` tambi칠n admite otros par치metros opcionales, como `orient`, `date_format`, `numpy`, `compression` y m치s. Podemos consultar la [documentaci칩n oficial](https://pandas.pydata.org/docs/reference/api/pandas.read_json.html) para obtener informaci칩n m치s detallada sobre estos par치metros y sus opciones.

````python
df_residuos = pd.read_json("residuos.json")
df_residuos.head()
````

## `pd.read_pickle`


**쯈u칠 es Pickle exactamente?**

Un archivo de tipo pickle es un archivo binario en Python que se utiliza para serializar y deserializar objetos,  su acr칩nimo viene de *Python Pickle Format* que fue desarrollado exclusivamente para Python. Leyendo solo esto podemos seguir sin tener claro que es un archivo de tipo pickle. Expliquemoslo de una forma m치s sencilla. Imagina que tienes un mont칩n de objetos especiales en tu habitaci칩n, como juguetes, libros, peluches, etc. Ahora, supongamos que quieres guardar todos estos objetos en una caja para poder llevarlos contigo o almacenarlos de manera segura.

Aqu칤 es donde entra en juego un archivo pickle. Es como una caja especial que te permite guardar todos estos objetos en ella. Pero, en lugar de objetos f칤sicos, el archivo pickle puede almacenar objetos digitales, como datos, n칰meros, listas, diccionarios y m치s.

El archivo pickle es como una caja m치gica que puede guardar todos tus objetos digitales de Python. Cuando guardas estos objetos en un archivo pickle, se convierten en una forma especial que se puede almacenar en tu computadora o enviar a alguien m치s.

La sintaxis b치sica de `pd.read_pickle` es la siguiente:

```python

```

Donde:

- `filepath_or_buffer`: Es la ruta del archivo pickle que deseas leer.

- `compression` (opcional): Es un par치metro opcional que indica el tipo de compresi칩n utilizado en el archivo pickle. Por defecto, es `'infer'`, lo que significa que la biblioteca intentar치 inferir autom치ticamente el tipo de compresi칩n. Puedes especificar un tipo de compresi칩n expl칤citamente, como `'gzip'` o `'bz2'`, si el archivo pickle est치 comprimido.
